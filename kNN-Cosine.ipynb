```python
import pandas as pd
import numpy as np

from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import NearestNeighbors
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
import matplotlib.lines as mlines
```

```python
df = pd.read_csv('CNAE-9.data', header = None)
```

```python
y = df.pop(0).values
```

```python
from sklearn.preprocessing import Normalizer
df_n = df.copy()
df_n.iloc[:,:] = Normalizer(norm='l2').fit_transform(df)
df_n.describe()
```

```python
NN_e = KNeighborsClassifier(metric = 'euclidean')
NN_c = KNeighborsClassifier(metric = 'cosine')
NN_m = KNeighborsClassifier(metric = 'manhattan')
NN_mink = KNeighborsClassifier(p=5)
```

```python
X = df.values
print("Not normalized \n")

scores = cross_val_score(NN_e, X, y, cv=10)
print("4x CV (Euclidean Distance):")
print(scores)
print(scores.mean())

scores = cross_val_score(NN_c, X, y, cv=10)
print("\n4x CV (Cosine):") 
print(scores)
print(scores.mean())

scores = cross_val_score(NN_m, X, y, cv=10)
print("\n4x CV (Manh):") 
print(scores)
print(scores.mean())


scores = cross_val_score(NN_mink, X, y, cv=10)
print("\n4x CV (Mink):") 
print(scores)
print(scores.mean())
```

```python
X = df_n.values
print("Normalized\n")

scores = cross_val_score(NN_e, X, y, cv=10)
print("4x CV (Euclidean Distance):")
print(scores)
print(scores.mean())

scores = cross_val_score(NN_c, X, y, cv=10)
print("\n4x CV (Cosine):") 
print(scores)
print(scores.mean())

scores = cross_val_score(NN_m, X, y, cv=10)
print("\n4x CV (Manh):") 
print(scores)
print(scores.mean())

scores = cross_val_score(NN_mink, X, y, cv=10)
print("\n4x CV (Mink):") 
print(scores)
print(scores.mean())
```
